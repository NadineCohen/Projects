```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
```

## Import and Merge Dataset

```python
vox2_spoof_df = pd.read_csv('vox2_p2_spoof_updated.csv')
vox2_df = pd.read_csv('vox2Updated.csv')
```

```python
merged_df = pd.merge(vox2_df, vox2_spoof_df, on = 'VoxCeleb2ID', how = 'left',validate = 'one_to_many')
```

```python
merged_df
```

```python
#merged_df[['Birthdate','Publish Date']] = pd.to_datetime(merged_df[['Birthdate','Publish Date']],format = '%Y-%m-%d')
merged_df['Birthdate'] = pd.to_datetime(merged_df['Birthdate'],format = '%Y-%m-%d')
merged_df['Publish Date'] = pd.to_datetime(merged_df['Publish Date'],format = '%Y-%m-%d')
```

```python
merged_df['Publish Age'] = merged_df['Publish Date'].dt.year - merged_df['Birthdate'].dt.year
```
